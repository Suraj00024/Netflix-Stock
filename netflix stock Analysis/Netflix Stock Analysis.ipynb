```python
import pandas as pd 
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv("Netflix.csv")
```

```python
df.head()
```

```python
sns.set(rc={"figure.figsize":(10,5)})
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date")
df.head()
```

# Volume of Stock Traded

```python
sns.lineplot(x="Date",y="Volume",data=df,label="Volume")
plt.title("Volume of Stock vs Time")
plt.show()
```

# Netflix Stock Price - High, Open, Close

```python
df.plot(y=["High","Close","Open"],title="Netflix Stock Price")
```

# Netflix Stock Price - Day, Month, Year

```python
fig,(ax,ax1,ax2) = plt.subplots(3,figsize = (15,10))
df.groupby(df.index.day).mean().plot(y="Volume",ax=ax,xlabel="Day")
df.groupby(df.index.month).mean().plot(y="Volume",ax=ax1,xlabel="Month")
df.groupby(df.index.year).mean().plot(y="Volume",ax=ax2,xlabel="Year")
```
